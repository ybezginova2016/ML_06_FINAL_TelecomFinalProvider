````python
# define ordinal encoding
ordinalencoder = OrdinalEncoder()

# transform data
result = ordinalencoder.fit_transform(data[['Type',
                                                     'PaperlessBilling', 
                                                     'PaymentMethod', 
                                                     'MonthlyCharges', 
                                                     'TotalCharges', 
                                                     'gender', 
                                                     'SeniorCitizen', 
                                                     'Partner',
                                                     'MultipleLines', 
                                                     'Dependents', 
                                                     'InternetService', 
                                                     'OnlineSecurity', 
                                                     'OnlineBackup', 
                                                     'DeviceProtection', 
                                                     'TechSupport', 
                                                     'StreamingTV', 
                                                     'StreamingMovies']])

result = pd.DataFrame(result, columns = ['Type',
                                                     'PaperlessBilling', 
                                                     'PaymentMethod', 
                                                     'MonthlyCharges', 
                                                     'TotalCharges', 
                                                     'gender', 
                                                     'SeniorCitizen', 
                                                     'Partner',
                                                     'MultipleLines', 
                                                     'Dependents', 
                                                     'InternetService', 
                                                     'OnlineSecurity', 
                                                     'OnlineBackup', 
                                                     'DeviceProtection', 
                                                     'TechSupport', 
                                                     'StreamingTV', 
                                                     'StreamingMovies'])

print(type(result))
result
````

````python
#  ,   
df_dropped = data.drop(['Type',
                                                     'PaperlessBilling', 
                                                     'PaymentMethod', 
                                                     'MonthlyCharges', 
                                                     'TotalCharges', 
                                                     'gender', 
                                                     'SeniorCitizen', 
                                                     'Partner',
                                                     'MultipleLines', 
                                                     'Dependents', 
                                                     'InternetService', 
                                                     'OnlineSecurity', 
                                                     'OnlineBackup', 
                                                     'DeviceProtection', 
                                                     'TechSupport', 
                                                     'StreamingTV', 
                                                     'StreamingMovies'], axis=1)

df_dropped
````
